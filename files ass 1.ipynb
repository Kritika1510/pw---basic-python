````python

1. What is an Exception in Python? Write the difference between Exceptions and Syntax Errors:
   - An exception in Python is an event that occurs during the execution of a program that disrupts the normal flow of instructions. This disruption is caused by an error of some kind. The main difference between exceptions and syntax errors is that syntax errors occur when the parser detects an incorrect statement syntax, while exceptions occur during execution when something unexpected happens.

2. What happens when an exception is not handled? Explain with an example:
   - When an exception is not handled, it propagates up the call stack until it's either caught by an appropriate exception handler or it reaches the top level of the program, resulting in termination. For example, if you try to open a file that doesn't exist without handling the `FileNotFoundError`, the program will crash.

3. Which Python statements are used to catch and handle exceptions? Explain with an example:
   - In Python, `try`, `except`, `else`, `finally`, and `raise` statements are used to catch and handle exceptions. For example:
     ```python
     try:
         x = int(input("Enter a number: "))
         y = 10 / x
     except ZeroDivisionError:
         print("Error: Division by zero!")
     else:
         print("Result:", y)
     ```

4. Explain with an example: try and else, finally, raise:
   - The `try` and `else` blocks are used together to specify a block of code to be executed if no exceptions occur. The `finally` block is used to specify a block of code to be executed regardless of whether an exception occurs or not. The `raise` statement is used to raise an exception manually. An example:
     ```python
     try:
         x = int(input("Enter a number: "))
         y = 10 / x
     except ZeroDivisionError:
         print("Error: Division by zero!")
     else:
         print("Result:", y)
     finally:
         print("This will always execute.")
     ```

5. What are Custom Exceptions in Python? Why do we need Custom Exceptions? Explain with an example:
   - Custom exceptions are user-defined exceptions that extend Python's built-in exception classes. We need them to create specialized exceptions that better represent specific error conditions in our programs. For example, let's say we're building a banking application and want to raise an exception when an account balance goes below zero:
     ```python
     class BalanceError(Exception):
         pass

     def withdraw(amount, balance):
         if amount > balance:
             raise BalanceError("Insufficient funds")
         return balance - amount

     try:
         balance = 100
         balance = withdraw(200, balance)
     except BalanceError as e:
         print("Error:", e)
     ```

6. Create a custom exception class. Use this class to handle an exception:
   - Here's how you can create a custom exception class and use it to handle an exception:
     ```python
     class CustomException(Exception):
         pass

     try:
         raise CustomException("This is a custom exception")
     except CustomException as e:
         print("Custom Exception Caught:", e)
     ```
````

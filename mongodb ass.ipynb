````python
Q1. 
   - MongoDB is a popular NoSQL database that stores data in a flexible, JSON-like format known as BSON. It is a document-oriented database, which means it stores data in collections of JSON-like documents.
   - Non-relational databases, or NoSQL databases, are databases that do not rely on the traditional tabular structures of relational databases. They offer flexibility in data storage and are designed to handle unstructured or semi-structured data.
   - MongoDB is preferred over SQL databases in scenarios where:
     - There is a need for flexible schema design, as MongoDB does not require a predefined schema.
     - There are large volumes of data with a variety of data types, as MongoDB can handle unstructured data efficiently.
     - Rapid development and scalability are required, as MongoDB is designed for horizontal scaling and can easily accommodate growing data needs.
     - There is a need for high availability and fault tolerance, as MongoDB supports replication and automatic failover.

Q2. 
   - **Flexible Schema**: MongoDB does not require a predefined schema, allowing for dynamic and flexible data models.
   - **Scalability**: MongoDB is horizontally scalable, meaning it can handle large volumes of data and traffic by distributing data across multiple servers.
   - **High Availability**: MongoDB supports replica sets, ensuring data availability and fault tolerance in case of server failures.
   - **Rich Query Language**: MongoDB provides a powerful query language with support for a wide range of query operations, including aggregation and geospatial queries.
   - **Indexing**: MongoDB supports indexing to improve query performance by efficiently retrieving data.
   - **GridFS**: MongoDB includes GridFS, a specification for storing and retrieving large files such as images, videos, and audio files.
   - **Aggregation Framework**: MongoDB provides an aggregation framework for performing data aggregation operations, such as grouping, sorting, and filtering.
   - **Geospatial Indexes**: MongoDB supports geospatial indexes and queries, allowing for location-based queries and analysis.

Q3. 
import pymongo

# Connect to MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")

# Create a database
mydb = client["mydatabase"]

# Create a collection
mycol = mydb["customers"]
```

Q4. **Using the database and the collection created in question number 3, write a code to insert one record, and insert many records. Use the find() and find_one() methods to print the inserted record:**
```python
# Insert one record
mydict = {"name": "John", "address": "Highway 37"}
insert_one_result = mycol.insert_one(mydict)

# Insert many records
mylist = [
  {"name": "Peter", "address": "Lowstreet 27"},
  {"name": "Amy", "address": "Apple st 652"},
  {"name": "Hannah", "address": "Mountain 21"},
  {"name": "Michael", "address": "Valley 345"},
  {"name": "Sandy", "address": "Ocean blvd 2"},
  {"name": "Betty", "address": "Green Grass 1"},
  {"name": "Richard", "address": "Sky st 331"},
  {"name": "Susan", "address": "One way 98"},
  {"name": "Vicky", "address": "Yellow Garden 2"},
  {"name": "Ben", "address": "Park Lane 38"},
  {"name": "William", "address": "Central st 954"},
  {"name": "Chuck", "address": "Main Road 989"},
  {"name": "Viola", "address": "Sideway 1633"}
]
insert_many_result = mycol.insert_many(mylist)

# Print inserted record using find_one()
print(mycol.find_one({"_id": insert_one_result.inserted_id}))

# Print all inserted records using find()
for x in mycol.find():
  print(x)
```

Q5. 
   - The `find()` method is used to query documents in a collection based on specified criteria.
   - Example:
     ```python
     # Find documents where the address starts with "S"
     query = {"address": {"$regex": "^S"}}
     result = mycol.find(query)
     
     for document in result:
         print(document)
     ```

Q6. 
   - Example:
     ```python
     # Sort documents by the "name" field in ascending order
     result = mycol.find().sort("name")
     
     for document in result:
         print(document)
     ```

Q7. 
   - `delete_one()`: Deletes the first document that matches the specified criteria.
   - `delete_many()`: Deletes all documents that match the specified criteria.
   - `drop()`: Deletes an entire collection from the database.
   - These methods are used to remove documents or collections from the database based on specified criteria. They help in managing data by allowing for selective deletion of records or entire collections.
````
